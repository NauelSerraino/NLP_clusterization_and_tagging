```python
from scipy import spatial


# # Define a function to get the vector representation of an expression
# def get_vector(expression, model):
#     vector = model.encode(expression)
#     return vector

def get_vector(word, model):
    try:
        vector = model.get_vector(word)
    except KeyError:
        try:
            words = word.split()
            vectors = [model.get_vector(w) for w in words]
            vector = np.mean(vectors, axis=0)
        except KeyError:
            vector = backup_model.encode(word)
    return vector



# Define the series of expressions
word_list = pd.read_pickle(f"{PATH_CHECKPOINTS}\corpus_inglese_grezzo.pkl")
word_list = word_list.str.lower().tolist()

# Compute the vectors for each expression
vectors = [get_vector(word, model) for word in word_list]

# Compute the cosine similarity distances between each vector and the vector for the word "man"
man_vector = get_vector('he', model=model)
woman_vector = get_vector('she', model=model)
person_vector = get_vector('it', model=model)

man_vector_backup = backup_model.encode('he')
woman_vector_backup = backup_model.encode('she')
person_vector_backup = backup_model.encode('it')

distances_man = []
for i in range(len(vectors)):
    try:
        distance = [1 - spatial.distance.cosine(vectors[i], man_vector)]
        distances_man.append(distance)
    except:
        distance = [1 - spatial.distance.cosine(vectors[i], man_vector_backup)]
        distances_man.append(distance)
        print(word_list[i])
        
distances_woman = []
for i in range(len(vectors)):
    try:
        distance = [1 - spatial.distance.cosine(vectors[i], woman_vector)]
        distances_woman.append(distance)
    except:
        distance = [1 - spatial.distance.cosine(vectors[i], woman_vector_backup)]
        distances_woman.append(distance)
        print(word_list[i])
        
distances_person = []
for i in range(len(vectors)):
    try:
        distance = [1 - spatial.distance.cosine(vectors[i], person_vector)]
        distances_person.append(distance)
    except:
        distance = [1 - spatial.distance.cosine(vectors[i], person_vector_backup)]
        distances_person.append(distance)
        print(word_list[i])

# Create a dictionary with the expression as the key and the distances as the values
results = pd.DataFrame({'MAN': distances_man, 'WOMAN': distances_woman
                        , "PERSON":distances_person
                        }, index=word_list)
```

```python
from scipy import spatial

# Define a function to get the vector representation of an expression
def get_vector(expression, model):
    vector = model.encode(expression)
    return vector

# Define the series of expressions
word_list = pd.read_csv(f"{PATH_CHECKPOINTS}\lista_parole_tradotte.csv")
expressions = word_list['Expressions']
# Compute the vectors for each expression
vectors = expressions.apply(get_vector, model=model)

# Compute the cosine similarity distances between each vector and the vector for the word "man"
man_vector = get_vector('he', model=model)
distances_man = vectors.apply(lambda x: 1 - spatial.distance.cosine(x, man_vector))

# Compute the cosine similarity distances between each vector and the vector for the word "woman"
woman_vector = get_vector('she', model=model)
distances_woman = vectors.apply(lambda x: 1 - spatial.distance.cosine(x, woman_vector))


# Create a dictionary with the expression as the key and the distances as the values
results = pd.DataFrame({'MAN': distances_man, 'WOMAN': distances_woman})
```

```python
pd.set_option("display.max_rows", None)
# Get the column name with the highest value for each row (excluding the 'list_of_words' column)
max_col = results[['MAN', 'WOMAN']].idxmax(axis=1)

# Extract the gender from the column name
associated_word = max_col.apply(lambda x: x.lower())

# Add the 'associated_word' column to the DataFrame
results['associated_word'] = associated_word
results["expressions"] = word_list["Expressions"]
```
